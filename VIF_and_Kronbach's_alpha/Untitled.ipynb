```python
def get_alpha_min(np_mat):
    cov = np_mat.cov()
    a_list = []
    for i in cov.columns:
        tmp = cov
        tmp = tmp.drop(i, axis=0)
        tmp = tmp.drop(i, axis=1).values
        leng = len(tmp)
        sum_v = sum([tmp[j, j] for j in range(leng)])
        v = sum_v/leng
        c = (sum(sum(tmp)) - sum_v)/(sum(range(leng))*2)
        a_list.append((i, leng*c/(v + (leng-1)*c)))
    
    return min(a_list, key = lambda t: t[1])
```
